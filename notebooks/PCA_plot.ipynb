```python
data = pd.read_csv('data/data.csv')
inputs = pd.read_csv('data/inputs.csv')
```

```python
inputs = inputs.fillna(0)
```

```python
data.rename(columns={'malate':'measured malate'}, inplace=True)
```

```python
data
```

```python
df = inputs.merge(data[['sample', 'measured malate']])
df.drop([6,7,8,9,10,11])
```

```python
df = df.set_index(df.columns[0])
correlation_matrix = df.corr()
```

```python
correlation_matrix
```

```python
X = df.loc[:, ['time', 'dilution', 'formate', 'NADH', 'glycine', 'malate', 'pyruvate',
       'serine', 'oxaloacetate', 'ATP', 'AMP', 'CoA', 'acetate', 'HCT', 'CO2',
       'acetyl-CoA', 'Fdh', 'Mdh', 'Pyc', 'TdcB']]
Y = df.loc[:, ["measured malate"]]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
PCA = PCA(n_components=2)
components = PCA.fit_transform(X)
PCA.components_
```

```python
componentsDf = pd.DataFrame(data = components, columns = ['PC1', 'PC2'])
pcaDf = pd.concat([componentsDf, Y], axis=1)

Y = data.loc[:, ["measured malate"]]

plt.figure(figsize=(12, 6))
sns.scatterplot(data=pcaDf, x="PC1", y="PC2", hue=Y.iloc[:, 0], 
                palette="magma",  # A palette where brighter colors represent higher values
                edgecolor=None)

# Show the plot
plt.show()
```
